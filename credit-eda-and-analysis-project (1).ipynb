# Análise exploratória de dados de crédito (EDA)

Notebook criado como parte do projeto proposto no curso de Análise de Dados da EBAC. O objetivo deste estudo é explorar um conjunto de dados relacionado a informações de crédito, buscando compreender padrões e características presentes nos dados através de técnicas de análise exploratória.

A tabela foi criada no AWS Athena junto com o S3 Bucket. O conjunto de dados utilizado nesta análise foi obtido a partir do repositório público disponível no GitHub: [credito.csv](https://github.com/andre-marcos-perez/ebac-course-utils/blob/main/dataset/credito.csv)


A base de dados original contém um volume maior de registros do que a amostra utilizada neste estudo. Para fins de aprendizado e execução mais eficiente das análises, foi utilizada apenas uma fração dos dados. Na prática, quanto maior a quantidade de dados analisados, maior tende a ser a confiabilidade das conclusões. No entanto, limitações computacionais ou financeiras podem tornar a redução do volume de dados uma estratégia comum em estudos exploratórios.

**Características dos clientes com maiores limites de crédito**

**Query:** 
select max(limite_credito) as limite_credito, escolaridade, tipo_cartao, sexo 
from credito
where escolaridade != 'na' and tipo_cartao != 'na' 
group by  escolaridade, tipo_cartao, sexo 
order by limite_credito desc
limit 10

![Valor limite](https://github.com/marianeneiva/sqlEBAC/blob/main/limite_desc.png?raw=true)

> Observa-se que a escolaridade não parece influenciar diretamente o limite concedido. O maior limite foi oferecido a um cliente sem escolaridade formal. Também não há relação clara entre o tipo de cartão e o limite. Entre os clientes com maiores limites, encontramos cartões das categorias gold, silver, platinum e blue, indicando que clientes de diferentes perfis financeiros podem receber limites elevados

**Características dos clientes com menores limites de crédito**

**Query:** 
select max(limite_credito) as limite_credito, escolaridade, tipo_cartao, sexo 
from credito
where escolaridade != 'na' and tipo_cartao != 'na' 
group by  escolaridade, tipo_cartao, sexo 
order by limite_credito asc


![Valor limite](https://github.com/marianeneiva/sqlEBAC/blob/main/limite_asc.png?raw=true)

>Nesta faixa de limite, não encontramos clientes com cartão platinum. A análise também mostra que a maioria desses clientes são mulheres, enquanto nos maiores limites predominam homens. Essa informação ajuda a compreender padrões de concessão de crédito e diferenças entre perfis de clientes, podendo orientar futuras decisões de marketing e política de limites.

**Comparativo de gastos entre homens e mulheres**

**Query:** 
select max(valor_transacoes_12m) as maior_valor_gasto, avg(valor_transacoes_12m) as  media_valor_gasto, min(valor_transacoes_12m) as min_valor_gasto, sexo 
from credito
group by sexo 


![Valor transacoes/sexo](https://github.com/marianeneiva/sqlEBAC/blob/main/quemgastamais.png?raw=true)

>Apesar de existirem diferenças nos limites de crédito, os valores gastos por homens e mulheres são bastante semelhantes. Isso indica que, em média, ambos os grupos movimentam quantias equivalentes, mesmo que os limites oferecidos pelo banco variem.

**Impacto do salário no limite de crédito**

**Query:** select avg(qtd_produtos) as qts_produtos, avg(valor_transacoes_12m) as media_valor_transacoes, avg(limite_credito) as media_limite,  sexo,   salario_anual from credito 
where salario_anual != 'na'
group by sexo, salario_anual
order by avg(valor_transacoes_12m) desc

![Valor salario_anualLimite](https://github.com/marianeneiva/sqlEBAC/blob/main/salarioAnualMediaLimite.png?raw=true)

>O resultado mostra que clientes com faixas salariais mais baixas tendem a ter limites de crédito menores. Isso evidencia que a renda impacta diretamente na quantidade de crédito disponível para cada cliente.

**Conclusão**

Conclusão

* A exploração inicial do dataset permitiu extrair diversos insights relevantes sobre os clientes:

* A maior parte dos clientes possui renda anual até 40K.

* A maior parte dos clientes é do sexo masculino.

* A escolaridade não apresenta influência clara sobre o limite de crédito ou o tipo de cartão.

* Clientes com maiores limites são, em sua maioria, homens.

* Clientes com menores limites são, em sua maioria, mulheres.

* Entre os menores limites, não há presença de cartão platinum.

* A faixa salarial impacta diretamente o limite de crédito disponível.

* Não existem clientes do sexo feminino com salário anual acima de 60K.

>Uma análise mais profunda pode ajudar a compreender melhor por que as mulheres tendem a ter limites menores, e esse padrão também pode refletir questões culturais que merecem atenção.
